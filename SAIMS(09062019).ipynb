# Nested for loops

```python
fNames = ['Nasir', 'Asad', 'Kamran']
lNames = ['Hussain', "Amin", 'Fazal']
for fName in fNames:
    for lName in lNames:
        print(fName,lName)
```

```python
table = int(input("Enter table number: "))

for num in range(1,11):
   
    print(f"{table} * {num} = {table*num}")
```

```python
for tables in range(2,6):
    print("##########")
    for num in range(1,11):
        
        print(f"{tables} * {num} = {tables*num}")
```
