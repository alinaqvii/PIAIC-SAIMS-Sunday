```python
def calculator(num1, num2, opr):
    if opr == '+':
        print(num1 + num2)
    elif opr == '-':
        print(num1 - num2)
    elif opr == '*':
        print(num1 * num2)
    elif opr == '/':
        print(num1 / num2)
    else:
        print('Wrong Operator')
```

```python
calculator(34,3,'-')
```

```python
#Discount Calculator
# If value is greater than 100 than add 10 in original value
# and if value is lesser than 100 than no discount
# 1% discount on value of 100 add 5 in original value
```

```python
def discount(x):
    if x > 100:
        y = x + 10
        return y
    elif x < 100:
        return x
```

```python
discount(120)
```

```python
discount(89)
```

```python
discount(100)
```
